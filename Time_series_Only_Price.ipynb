```python
def adfuller_test(price):
    result=adfuller(price)
    labels = ['ADF Test Statistic','p-value','#Lags Used','Number of Observations']
    for value,label in zip(result,labels):
        print(label+' : '+str(value) )
        
        if result[1] <= 0.05:
            print("strong evidence against the null hypothesis(Ho), reject the null hypothesis. Data is stationary")
        else:
            print("weak evidence against null hypothesis,indicating it is non-stationary ")


```

H0: The null hypothesis: It is a statement about the population that either is believed to be true or is used to put forth an argument unless it can be shown to be incorrect beyond a reasonable doubt.

H1: The alternative hypothesis: It is a claim about the population that is contradictory to H0 and what we conclude when we reject H0.

#Ho: It is non-stationary
#H1: It is stationary

```python
adfuller_test(Pedralbes['price_month(€)'])
```

->>> High p-values indicate that your evidence is not strong enough to suggest an effect exists in the population. An effect might exist but it's possible that the effect size is too small, the sample size is too small, or there is too much variability for the hypothesis test to detect it.

->> Since the Ho was rejected, the alternative hypothesis (Ha) could be accepted

Autocorrelation represents the degree of similarity between a given time series and a lagged version of itself over successive time intervals.

1.An autocorrelation of +1 represents a perfect positive correlation, while an autocorrelation of negative 1 represents a perfect negative correlation.

#### Changed Pedralbes with only quarter +price to predict the price changes w/o other features first 

```python
Pedralbes=df[['quarter','price_month(€)']]
Pedralbes
```

```python
# Split data into train / test sets 
train = Pedralbes.iloc[:len(Pedralbes)-17] 
test = Pedralbes.iloc[len(Pedralbes)-17:] # there are 22 rows (17 rows for train/5 rows for testing)
  
# Fit a SARIMAX (0,0,0)(0,1,0)[12]   on the training set 
 
from statsmodels.tsa.statespace.sarimax import SARIMAX 
  
model = SARIMAX(train['price_month(€)'],  
                order = (0, 0, 0),  
                seasonal_order =(0, 1, 0, 12)) 
  
result = model.fit() 
result.summary() 
```

```python
start = len(train) 
end = len(train) + len(test) - 1
  
# Predictions for one-year against the test set 
predictions = result.predict(start, end, 
                             typ = 'levels').rename("Predictions") 
  
# plot predictions and actual values 
predictions.plot(legend = True) 
test['price_month(€)'].plot(legend = True) 
```

```python
from sklearn.metrics import mean_squared_error 
from statsmodels.tools.eval_measures import rmse 
  
# Calculate root mean squared error 
rmse(test['price_month(€)'], predictions) 
  
# Calculate mean squared error 
mean_squared_error(test['price_month(€)'], predictions) 

```

```python
# Train the model on the full dataset 
model = model = SARIMAX(Pedralbes['price_month(€)'],  
                      order = (0, 0, 0),  
                seasonal_order =(0, 1, 0, 12)) 
result = model.fit() 
  
# Forecast for the next 6 quarters
forecast = result.predict(start = len(Pedralbes),  
                          end = (len(Pedralbes)-1) + 6,  
                          typ = 'levels').rename('Forecast') 
  
# Plot the forecast values 
Pedralbes['price_month(€)'].plot(figsize = (18, 7), legend = True) 
forecast.plot(legend = True) 
```
